```python
plt.figure(figsize=(200,40))
ax = sns.lineplot(x="Country", y="Deaths", data=df_coro)
plt.ylabel('Deaths', size=180)
plt.title('Death Case in Various Countries', size=180)
plt.xticks(size=50,rotation = 120)
plt.yticks(size=140, rotation=240)
sns.despine(left=True);
```
