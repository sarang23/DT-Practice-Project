Now removing outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(dfcor))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
df_new=dfcor[(z<3).all(axis=1)]
```

```python
df_new
```

```python
df_newavo=df_new.drop(['Unnamed: 0','Year'],axis=1)
```

Now we can Drop Unnamed and Year Column to make data Clean and make a new Data frame to conclude

```python
#Now we devide the Data into input and output
df_x=df_newavo.drop(columns=["AveragePrice"])
y=df_newavo[["AveragePrice"]]
```
