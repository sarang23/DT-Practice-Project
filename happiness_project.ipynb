```python
import matplotlib.pyplot as plt 
from pandas.plotting import scatter_matrix
```

```python
features=['Country', 'Region', 'Happiness Rank', 'Happiness Score',
       'Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']
```

```python
scatter_matrix(df_happy[features],alpha=1, figsize=(40,20), ax=None, grid=False, diagonal='hist', marker='.', density_kwds=None, hist_kwds=None, range_padding=0.05) 
```

```python
df_happy.corr().round(2)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_happy.corr().round(2),annot=True,linewidth=1,linecolor="black",fmt='10')
```
