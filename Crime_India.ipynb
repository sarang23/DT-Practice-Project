# Crime in India 



We will Consider 10_Property_stolen_and_recovered.csv for our Study.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_india=pd.read_csv("10_Property_stolen_and_recovered.csv")
df_india
```

```python
df_india.isnull()
```

```python
df_india.isna().sum()
```

Here We have Found No null Values in the column.

```python
df_india.describe()
```

```python
df_india.hist(figsize=(10,10),edgecolor='c')
plt.tight_layout()
plt.show()
```

```python
df_india.info()
```

```python
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Area_Name", y="Cases_Property_Stolen", data=df_india,errwidth=None)
plt.ylabel('Cases_Property_Stolen', size=18)
plt.title('Stolen Property vs Various State', size=18)
plt.xticks(size=12,rotation = 60)
plt.yticks(size=14)
sns.despine(left=True);
```

Now we will See the Stolen Property Over various Years. We have Observed that Maharashtra has the highest No. of Stolen Cases

```python
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Year", y="Cases_Property_Stolen", data=df_india,errwidth=None)
plt.ylabel('Cases_Property_Stolen', size=18)
plt.title('SYears vs Various State', size=18)
plt.xticks(size=12,rotation = 60)
plt.yticks(size=14)
sns.despine(left=True);
```

In Year 2010 There were highest No. of Stolen Cases

Now we Can consider for Stolen cases and recovered Cases for this we can do scatter graph to see the coorelation between these two entities.

```python
sns.scatterplot(x=df_india.Cases_Property_Recovered, y=df_india.Cases_Property_Stolen)
```

This is Showing positive corellation.

```python
x=df_india.Cases_Property_Recovered
y=df_india.Cases_Property_Stolen
```
